Model Evaluation & Interpretation

To provide a clear foundation for the analysis of our models, we first examine the general strengths and weaknesses of both models. This helps demonstrate why they were chosen as primary candidates for analyzing the given data set about falls in hospitals.

General strengths and weaknesses of:

Logistic Regression

Advantages:
•	Simplicity: Very easy to implement and interpret, especially for small data sets
•	Efficiency: Fast and computationally efficient, well suited for structured data sets
•	Interpretability: The coefficients are easy to understand and directly show the influence of the features
•	Regularization: Can be extended with L1 or L2 regularization to avoid overfitting

Disadvantages:
•	Limited to linear relationships: Can only model linear relationships, which is not sufficient for more complex problems
•	Feature scaling: Works better if the features are scaled
•	Not good for non-linear data: Performance can suffer with complex or high-dimensional data

XGBClassifier (Gradient Boosting of Decision Trees)

Advantages:
•	Complexity handling: Very well suited for complex and non-linear problems.
•	Feature engineering: Can work with non-normalized or non-scaled data.
•	Flexibility: Can be highly customized with hyperparameter tuning, such as max_depth, learning_rate, and n_estimators.
•	Efficiency: Optimized for high speed and performance, especially on large datasets.
•	Feature Importance: Provides information about which features influence the prediction.

Disadvantages:
•	Complexity: Higher complexity compared to logistic regression, both in implementation and interpretation.
•	Computing effort: More computationally intensive and requires more memory resources.
•	Susceptible to overfitting: Especially if not properly regularized or with small data sets

Now that the advantages and disadvantages of our two models have been outlined, we would like to find out which model is particularly advantageous for which stakeholder group of a hospital and under which circumstances the respective model is preferable (in everyday hospital life).
Logistic regression is preferrable when:
•	There are limited Computational Resources: Logistic Regression is computationally efficient and works in environments with constrained resources
•	Stakeholders Require Simplicity: Stakeholders benefit from the straightforward coefficients and odds ratios provided by this model, allowing them to understand and justify predictions.
•	Linear Relationships Exist: When the relationship between predictors (e.g., age, medications) and the outcome (fall risk) is approximately linear.
Stakeholder target groups for this model would primarily be people and groups who are not familiar with statistical models and who need results from models quickly. So we are talking about doctors, clinicians, hospital administrators, or decision makers.
The model could be useful for the following applications in everyday clinical practice:
•	Risk Scoring: Logistic Regression can generate simple fall-risk or other risk scores for patients, which clinicians can easily interpret. A sepsis risk score for example, using data like vital signs (heart rate, blood pressure) and lab results (white blood cell count, lactate levels), helps prioritize patients for early intervention, reducing complications
•	Decision Support: Outputs like probabilities and feature coefficients can be integrated into electronic health record (EHR) systems to provide alerts (e.g., "Patient X has a 70% likelihood of falling"). Staffing plans could be adapted accordingly for higher-risk groups of people to improve care and reduce risks
•	Educational Tool: As it's transparent, it can be used in staff training to understand key risk factors for falls or other relevant metrics. Staff training could be more specific to real day-to-day problems/situations and therefore more useful than general training on all medically possible cases

XGB-Classifier is preferrable when:
•	Complex and Non-Linear Relationships exist: When interactions between variables (e.g., age, mobility, medications) are not straightforward or linear.
•	Larger, High-Dimensional Datasets need to be analyzed: It handles large datasets with many features efficiently, allowing for complex analysis in diverse patient populations.
•	Maximizing Prediction Accuracy is necessary: XGB-Classifier often outperforms simpler models in terms of predictive power, especially with thorough hyperparameter tuning
Stakeholder target groups for this model would primarily be people and groups who are familiar with statistical models and are more interested in the most accurate relations of variables in specific datasets. We are therefore talking about researchers, management and administrative staff such as the members of the Sales Department. The latter must, for example, accurately assess whether future investments or acquisitions are really worthwhile.
The model could be useful for the following applications in everyday clinical practice:
•	Identifying Multi-Factorial Fall Risks
While logistic regression can identify general fall risk factors, XGB-Classifier can analyze complex interactions between multiple variables, such as age, medication type, comorbidities, environmental factors (like ward type), and activity logs. Therefore it is possible to use XGB-Classifier to generate granular fall-risk predictions for patients with unique, multi-dimensional profiles. Pair these predictions with intervention strategies that dynamically adjust based on real-time patient data to improve patient care
•	Prioritizing High-Risk Surgeries
Unlike logistic regression, XGB-Classifier can handle non-linear interactions between risk factors for surgical outcomes (e.g., age, prior health conditions, or lab test abnormalities). This would allow to develop a pre-operative risk assessment tool to classify patients by their likelihood of complications. Surgeons could use this tool to prioritize monitoring resources or adjust surgery schedules
•	Preventing Adverse Drug Reactions
Interactions between medications, genetic factors, and pre-existing conditions can lead to adverse drug reactions, which are difficult to model with linear techniques. Therefore use XGB-Classifier to predict the likelihood of adverse reactions based on patient pharmacogenomic profiles, drug combinations, and health records. Pharmacists can implement alternative medication plans or monitor flagged patients closely.
•	Prioritize Equipment Investments
The XGB-Classifiers ability to handle complex relationships and large datasets makes it well-suited for evaluating multi-faceted scenarios. It helps to justify the need for additional equipment purchases in high-risk areas. It supports the development of a persuasive business case for hospital management or external funders and offers data-driven recommendations for the most effective deployment of resources

Practical Challenges in Implementation
Data Challenges:
•	Incomplete Data: Missing or incomplete patient records may skew predictions.
o	Solution: Use imputation methods (e.g., mean, median) or specialized algorithms like k-NN imputation
o	For our given Dataset we used insert
•	Imbalanced Classes: In our dataset non-fall cases were rare compared to fall cases.
o	Solution: Apply techniques like oversampling (SMOTE) or assign class weights.
Operational Challenges:
•	Integration with Clinical Systems:
o	Ensure the model is embedded in EHR systems for real-time predictions
o	Train clinicians to interpret and respond to model outputs effectively
•	Regulatory Compliance:
o	Address data privacy and security, adhering to frameworks like GDPR or HIPAA
•	Resistance to Change: Stakeholders may be hesitant to rely on AI-driven models over traditional clinical judgment
Stakeholder Communication:
•	Simplify explanations for clinical teams with minimal technical knowledge so that they can work with the model output
•	Provide visualization tools to present results (e.g., ROC curves, SHAP plots)
Bias in Data:
•	Models trained on biased data (e.g., underrepresented groups) may lead to inequitable predictions.
Model Maintenance Challenges:
•	Updating Models: Clinical guidelines and patient demographics change over time, requiring periodic retraining of models.
•	Monitoring Performance: Continuous evaluation is needed to ensure that the model performs well in real-world settings

Conclusion:
•	Logistic Regression: Easy to implement, interpret, and deploy for immediate clinical use.
•	XGBClassifier: Offers superior predictive accuracy and actionable insights for high-complexity data but requires additional effort for integration and explanation.

Comparison to Classical Fall Risk Factors in the Literature
These are the most common fall risk factors in no order of relevance in the Literature:
1.	Age: Older patients are at higher risk.
2.	Medication Use: Sedatives, antihypertensives, and polypharmacy.
3.	History of Falls: One of the strongest predictors of future falls.
4.	Mobility Impairments: Difficulty walking or using assistive devices.
5.	Cognitive Decline: Dementia or delirium increases risk.
6.	Chronic Illnesses: Such as Parkinson’s disease or stroke.
7.	Environmental Hazards: Poor lighting, slippery floors, or cluttered spaces.
8.	Visual Impairments: Decreased vision can affect mobility and balance.
9.	Dizziness or Balance Disorders: Often related to vestibular or circulatory issues.
